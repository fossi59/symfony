- pour choisir l'algorythme qui va predire le prix d'une maison à partir de sa surface, on installera la librairie "scikit-learn"

```
!pip install scikit-learn
```

- choisir l'algorythme approprié à notre probléme

```
from sklearn import linear_model
```

- choisir modele regression linéaire

```
model = linear_model.LinearRegression()
```

on vient de créer l'algorythme de la regression linéaire

Entrainement de l'algorythme

```
y = df['price']
x = df.drop(columns=['price'])
```

```
x
```

```
y
```

```
model.fit(x,y)
```

- le model est entrainé
- etape suivante est de tester la fiabilité du model

```
predicted = model.predict(x)
```

```
plp.figure(figsize=(10,5))
plp.xlabel('area')
plp.ylabel('price')
plp.title('Linear regression')
plp.scatter(df['area'],df['price'], color='red', marker='+')
plp.plot(x,predicted, color='blue')
```
